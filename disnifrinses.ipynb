```python
# Count occurrences of each hair color
hair_color_counts = disney['HairColor'].value_counts()

# Define a list of colors for the bars
colors = ['#670380', '#821e9c', '#ba35db', '#c079d1', '#eabcf5']

# Create a bar chart
plt.figure(figsize=(10, 6))
plt.bar(hair_color_counts.index, hair_color_counts.values, color=colors[:len(hair_color_counts)])
plt.title('Hair Color Distribution of Disney Princesses')
plt.xlabel('Hair Color')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
