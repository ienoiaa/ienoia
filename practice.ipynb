```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

dsny = pd.read_csv('/content/disney_princess_popularity_dataset_300_rows.csv')
display(dsny)
```

```python
pop_score = dsny['PopularityScore'].value_counts()

plt.figure(figsize=(10, 6))  # Adjust figure size as needed
plt.bar(range(len(dsny)), dsny['PopularityScore'], color='violet')
plt.xlabel("Row Index")
plt.ylabel("Popularity Score")
plt.title("Popularity Scores for All 300 Rows")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
pop_score = dsny['MovieRuntimeMinutes'].value_counts()

plt.figure(figsize=(10, 6))  # Adjust figure size as needed
plt.bar(range(len(dsny)), dsny['MovieRuntimeMinutes'], color='pink')
plt.xlabel("Row Index")
plt.ylabel("Movie Runtime by Minutes")
plt.title("Movie Runtime for All 300 Rows")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
princess_counts = dsny['PrincessName'].value_counts()
percentages = (princess_counts / 300) * 100
plt.figure(figsize=(8, 8))
plt.pie(percentages, labels=percentages.index, autopct='%1.1f%%', startangle=90,  colors=sns.color_palette('Set3'))
plt.title("Distribution of Disney Princess Names")
plt.show()
```
